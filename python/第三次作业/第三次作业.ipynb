```python
import matplotlib.pyplot as plt
# 支持中文
plt.rcParams['font.family'] = ['sans-serif']
plt.rcParams['font.sans-serif'] = ['SimHei']
# 结局负数乱码
plt.rcParams['axes.unicode_minus']=False
import numpy as np
```

```python
"""
使用numpy与matplotlib在jupyter中绘制正弦与余弦曲线,
并显示图例
"""
```

# 这是我画的，另外两个是用gpt画着玩的

```python
x = np.linspace(-4, 4, 200)
y1 = np.sin(x)
y2 = np.cos(x)
plt.figure(1)
plt.legend(['正弦','余弦'])
#获取当前axes
ax = plt.gca()
#坐标系中right,top不显示
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
#设置ticks显示位置
ax.xaxis.set_ticks_position('bottom') 
ax.yaxis.set_ticks_position('left')
#移动坐标轴位置：中心位置为(0,0)
ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))

# 绘制曲线并添加标签用于图例
plt.plot(x, y1, label='正弦')
plt.plot(x, y2, label='余弦')

# 添加图例要在绘制曲线之后
plt.legend(loc='upper left')

plt.show()
```

```python
# 准备数据
x = np.linspace(-4, 4, 200)
y1 = np.sin(x)
y2 = np.cos(x)

# 创建图形和轴
fig, ax = plt.subplots()

# 设置图形的背景色
ax.set_facecolor('#f0f0f0')

# 隐藏右边和上边的轴线
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# 设置x轴和y轴的ticks显示在底部和左边
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# 移动左边和底部轴线到数据坐标(0, 0)并设置轴线颜色
ax.spines['left'].set_position(('data', 0))
ax.spines['left'].set_color('gray')
ax.spines['bottom'].set_position(('data', 0))
ax.spines['bottom'].set_color('gray')

# 绘制图形，使用更吸引的样式
ax.plot(x, y1, label="$sin(x)$", color="#0077b6", linewidth=2.5, linestyle="-", marker='o', markevery=20, markersize=6)
ax.plot(x, y2, label="$cos(x)$", color="#f4a261", linewidth=2.5, linestyle="--", marker='x', markevery=20, markersize=6)

# 添加图例，并改善图例的外观
leg = ax.legend(frameon=True, loc='upper left', fontsize=10)
leg.get_frame().set_edgecolor('black')
leg.get_frame().set_facecolor('#f0f0f0')

# 添加标题
ax.set_title('Trigonometric Functions', fontsize=18, fontweight='bold', color='navy')

# 添加X和Y轴标签并设置样式
ax.set_xlabel('X-Axis', fontsize=12, fontweight='bold', color='darkgreen')
ax.set_ylabel('Y-Axis', fontsize=12, fontweight='bold', color='darkgreen')

# 添加细节的网格并设置样式
ax.grid(True, linestyle='--', linewidth=1, alpha=0.9, color='grey')

# 设置坐标轴范围
ax.set_xlim(-4, 4)
ax.set_ylim(-1.5, 1.5)

# 设置坐标轴刻度标签的大小和颜色
ax.tick_params(axis='both', which='major', labelsize=10, colors='blue')

# 修改x和y的主要刻度线，使其更长更突出
ax.tick_params(axis='both', which='major', length=7, width=2)

# 给特殊点添加注释
ax.annotate('Origin', xy=(0, 0), xytext=(1, -1), 
            arrowprops=dict(facecolor='black', shrink=0.05), fontsize=12, color='red')

# 保存图形到文件
# plt.savefig('trigonometric_functions.png', dpi=300)

# 显示图形
plt.show()
```

```python
from matplotlib.patches import ArrowStyle
import matplotlib.pyplot as plt
import numpy as np

# 准备数据
x = np.linspace(-4, 4, 200)
y1 = np.sin(x)
y2 = np.cos(x)

# 创建图形和轴
fig, ax = plt.subplots()

# 设置图形的背景色
ax.set_facecolor('#e0f7fa')  # 设置为淡青色背景

# 隐藏右边和上边的轴线
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# 设置x轴和y轴的ticks显示在底部和左边
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# 移动左边和底部轴线到数据坐标(0, 0)
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')

# 设置轴线颜色和样式
ax.spines['left'].set_color('gray')
ax.spines['bottom'].set_color('gray')
ax.spines['left'].set_linewidth(0.5)
ax.spines['bottom'].set_linewidth(0.5)

# 高亮x轴和y轴的0点刻度
ax.tick_params(axis='both', direction='inout', length=6, color='b', grid_alpha=0.5)

# 绘制y1和y2
line1, = ax.plot(x, y1, label="$\sin(x)$", color="#42a5f5", linewidth=2, linestyle="-")
line2, = ax.plot(x, y2, label="$\cos(x)$", color="#ef5350", linewidth=2, linestyle="--")

# 为y1和y2增加标记
ax.scatter(x[::20], y1[::20], c=line1.get_color(), s=50, edgecolor='w', zorder=3)
ax.scatter(x[::20], y2[::20], c=line2.get_color(), s=50, edgecolor='w', zorder=3)

# 添加图例，并改善图例的外观
leg = ax.legend(frameon=True, loc='upper left', fontsize=9, edgecolor='black')
leg.get_frame().set_facecolor('#e0e0ff')  # 设置为淡紫色背景
leg.get_frame().set_alpha(1)

# 添加标题
ax.set_title('Trigonometric Functions - Sine and Cosine', fontsize=15, fontweight='bold', color='navy')

# 添加X和Y轴标签
ax.set_xlabel('X axis', fontsize=12, fontweight='bold', color='darkblue')
ax.set_ylabel('Y axis', fontsize=12, fontweight='bold', color='darkblue')

# 添加更详细的网格
ax.grid(True, 'both', 'both', color='gray', linewidth=0.5, linestyle='--', alpha=0.3)

# 设置坐标轴范围
ax.set_xlim(-4.5, 4.5)
ax.set_ylim(-1.5, 1.5)

# 显示特殊点的注释
ax.annotate('Origin', xy=(0, 0), xytext=(-0.5, 0.5), textcoords="offset points",
            arrowprops=dict(arrowstyle='->', fc='black', shrinkA=0, shrinkB=0),
            ha='center', fontsize=10, color='red', fontweight='bold')

# 展示图形
plt.show()
```
